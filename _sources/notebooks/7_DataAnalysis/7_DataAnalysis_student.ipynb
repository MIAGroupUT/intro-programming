# Data analysis

:::{admonition} Learning goals
:class: note
After finishing this chapter, you are expected to
* import data in CSV or Excel format
* process this data
* write data to different file formats
* apply simple filters to images
:::

In this chapter, we're going to be working with a few commonly used packages for data. 

## Scipy
According to it's website, the [SciPy](https://scipy.org/) package does the following:

> SciPy provides algorithms for optimization, integration, interpolation, eigenvalue problems, algebraic equations, differential equations, statistics and many other classes of problems.

You'll see that with NumPy and SciPy, you are able to replicate a lot of the functionality that is present in Matlab. Scipy even offers [utilities to load and save Matlab files](https://docs.scipy.org/doc/scipy/reference/io.matlab.html). 

## Image analysis
The [scikit-image](https://scikit-image.org/) package offers a lot of functionality in image analysis.

## Advanced visualization
You have previously worked with Matplotlib to generate plots. Matplotlib is by far the most widely used visualization package in Python, but there are other - more specialized - packages available if you want to make other figures. Here, we'll touch upon two well-developed packages: [Seaborn](https://seaborn.pydata.org/) and [PyVista](https://docs.pyvista.org/version/stable/). As often in this course, we aim to introduce some common functionality and a quick introduction these packages, but encourage you to take a close look at their documentation to find out the broad range of functionality that they offer.

### Seaborn
Seaborn is a data package that is very useful for visualization of data. It is built on top of Matplotlib and uses many of the same objects.
